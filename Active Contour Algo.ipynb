```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
from PIL import Image
from numpy import asarray
#grayscaling using matplotlib and the formula Y = 0.2126R + 0.7152G + 0.0722B


#function that splits the image into an array of pixel values and takes the dot product with the correct grayscaling values to return the grayscaled image
def rgb2graysc(rgb):
    return np.dot(rgb[...,:3], [0.2126, 0.7152, 0.0722])
#imports image
img = Image.open('/content/Intersection.jpg')
img.convert('RGB')
img_data = asarray(img)
plt.imshow(img)
plt.show()
gray = rgb2graysc(img_data)
#displays the grayscaled image
#grayscaled = Image.fromarray(gray)
#grayscaled.save('greyscale.png')

plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin = 0, vmax = 255)
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour

s = np.linspace(0, 2*np.pi, 400)
x = 220 + 100*np.cos(s)
y = 100 + 100*np.sin(s)
init = np.array([x, y]).T


cv2.imread(gray)
image = cv2.Canny(gray,100,200)

snake = active_contour(image, init, alpha=0.015, beta = 0.2, gamma = 0.001)

fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(image, cmap=plt.cm.gray)
ax.plot(init[:, 0], init[:, 1], '--r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, image.shape[1], image.shape[0], 0])

plt.show()
```

```python
#CV2 Active Contour for Comparision
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/Intersec (2).jpg')

img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2_imshow(img_gray)

canny_im = cv2.Canny(img_gray,150,255)
cv2_imshow(canny_im)

contours, hierarchy = cv2.findContours(image=canny_im,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image=image,contours=contours,contourIdx=-1,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(image)
```

```python
ret, thresh_im = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(image=thresh_im,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image=image,contours=contours,contourIdx=-1,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(image)
```
