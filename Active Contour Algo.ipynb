```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
from PIL import Image
from numpy import asarray
#grayscaling using matplotlib and the formula Y = 0.2126R + 0.7152G + 0.0722B


#function that splits the image into an array of pixel values and takes the dot product with the correct grayscaling values to return the grayscaled image
def rgb2graysc(rgb):
    return np.dot(rgb[...,:3], [0.2126, 0.7152, 0.0722])
#imports image
img = Image.open('/content/Intersection.jpg')
img.convert('RGB')
img_data = asarray(img)
plt.imshow(img)
plt.show()
gray = rgb2graysc(img_data)

#displays the grayscaled image
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin = 0, vmax = 255)
plt.show()
```

```python
#CV2 Active Contour for Comparision
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/Intersection.jpg')

img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2_imshow(img_gray)

canny_im = cv2.Canny(img_gray,100,200)
cv2_imshow(canny_im)

contours, hierarchy = cv2.findContours(image=canny_im,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image=image,contours=contours,contourIdx=-1,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(image)
```

```python
ret, thresh_im = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(image=thresh_im,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

cv2.drawContours(image=image,contours=contours,contourIdx=-1,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(image)
```
